```python
import torch
```

```python
N=torch.zeros((27,27,27),dtype=torch.int32)
N.shape
```

```python
chars=sorted(list(set(''.join(words))))
stoi ={s:i+1 for i,s in enumerate(chars)}
stoi['.']=0
itos = {i:s for s,i in stoi.items()}
```

```python
for w in words:
    chs=['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1=stoi[ch1]
        ix2=stoi[ch2]
        ix3=stoi[ch3]
        N[ix1,ix2,ix3]+=1
```

```python
N[0]
```

```python
p=N[0].float()
p=p/p.sum()
p
```

```python
P = (N+1).float()
P/=P.sum(1, keepdim=True)
```

```python
P[1].sum()
```

```python
g=torch.Generator().manual_seed(2147483647)
for i in range(5):
    for j in range(i):
        out=[]
        ix=0
        while(True):
            p=P[ix][j]
            #p=N[ix].float()
            #p/=p.sum()
            ix=torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
            out.append(itos[ix])
            if ix==0:
                break
        print(''.join(out))
```
